```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year,df['per capita income (US$)'],color='red',marker='+')
```

```python
new_df = df.drop('per capita income (US$)',axis='columns')
new_df
```

```python
perincome = df['per capita income (US$)']
perincome
```

```python
reg= linear_model.LinearRegression()     
reg.fit(new_df,perincome)
```

```python
reg.predict([[2024]])
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year,df['per capita income (US$)'],color='red',marker='+')
plt.plot(df.year,reg.predict(new_df),color= 'blue')
```
