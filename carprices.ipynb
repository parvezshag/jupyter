```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("carprices.csv")
df
```

```python
put_dummies = pd.get_dummies(df['Car Model'],dtype=int)
put_dummies
```

```python
merged = pd.concat([df,put_dummies],axis='columns')
merged
```

```python
final = merged.drop(['Car Model','Mercedez Benz C class'],axis='columns')
final
```

```python
X = final.drop('Sell Price($)', axis='columns')
X
```

```python
y = final['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict(X)
```

```python
model.score(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[45000,4,0,0]]) #Price of mercedez benz that is 4 yr old with mileage 45000
```

```python
model.predict([[86000,7,0,1]]) #Price of BMW X5 that is 7 yr old with mileage 86000
```

```python
%matplotlib inline
plt.xlabel('Age')
plt.ylabel('Price(US $)')
plt.scatter(df['Age(yrs)'],df['Sell Price($)'],color='red',marker='+')
plt.plot(df['Age(yrs)'],model.predict(X),color= 'blue')
```
